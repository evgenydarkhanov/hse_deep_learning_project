### **поиск `BM25`**

- [BM25](https://github.com/dorianbrown/rank_bm25)
- Note that this package doesn't do any text preprocessing. If you want to do things like lowercasing, stopword removal, stemming, etc, you need to do it yourself.
- The only requirements is that the class receives a list of lists of strings, which are the document tokens.
- Good to note that we also need to tokenize our query, and apply the same preprocessing steps we did to the documents in order to have an apples-to-apples comparison

```python
%%capture
!pip install rank_bm25
!pip install pymystem3
!pip install pymorphy2==0.8 -i
```

**quick start**

```python
from rank_bm25 import BM25Okapi
```

```python
corpus = [
    "Hello there good man!",
    "It is quite windy in London",
    "How is the weather today?"
]

tokenized_corpus = [doc.split(" ") for doc in corpus]

bm25 = BM25Okapi(tokenized_corpus)
print(bm25)
# <rank_bm25.BM25Okapi at 0x1047881d0>

query = "windy London"
tokenized_query = query.split(" ")

doc_scores = bm25.get_scores(tokenized_query)
print(doc_scores)
# array([0.        , 0.93729472, 0.        ])

bm25.get_top_n(tokenized_query, corpus, n=1)
# ['It is quite windy in London']
```

```python
TOKEN_WORD = re.compile('[а-яА-ЯёЁ]{2,}')

def remove_lemmatize_preprocessing(sentence: str, words_to_remove: set) -> list:
    """ приводит к нижнему регистру, убирает числа и мусорные слова, лемматизирует """
    if isinstance(sentence, str):

        regex_tmp = re.findall(TOKEN_WORD, sentence.lower())
        remove_tmp = [lemmatizer.parse(token)[0].normal_form for token in regex_tmp if token not in words_to_remove]
        return remove_tmp
    return []


def one_sentence_preprocessing(sentence: str, words_to_remove: set) -> str:
    """ применяет препроцессинги к одному предложению """
    sentence_words = remove_lemmatize_preprocessing(sentence, words_to_remove)
    result = ' '.join(sentence_words)
    return result
```

```python
services_cleaned = services.apply(lambda cell: one_sentence_preprocessing(cell, stop_words))
services_tokenized = services_cleaned.apply(lambda cell: cell.split())
services_tokenized
```

```python
bm25_vk = BM25Okapi(services_tokenized)


def make_query_bm_25(query: str,
                     corpus=services,
                     words_to_remove: set = stop_words,
                     top: int = 1):

    query_cleaned = one_sentence_preprocessing(query, words_to_remove)
    query_tokenized = query_cleaned.split()
    result = bm25_vk.get_top_n(query_tokenized, corpus, n=top)

    return result
```

```python
for query in vk_queries:
    print(f"QUERY: {query}")
    result = make_query_bm_25(query, top=2)
    for elem in result:
        print(f"{elem}\n")
    print('====================================')
```

```python
rubert_tiny_path = "/home/darhanovev/VK RAG/rubert_dir"

tokenizer = AutoTokenizer.from_pretrained(rubert_tiny_path)
model = AutoModel.from_pretrained(rubert_tiny_path)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
def embed_bert_cls(text, model, tokenizer):
    t = tokenizer(text, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**{k: v.to(model.device) for k, v in t.items()})
    embeddings = model_output.last_hidden_state[:, 0, :]
    embeddings = torch.nn.functional.normalize(embeddings)
    return embeddings[0].cpu().numpy()
```

```python
services_emb = np.array([embed_bert_cls(service, model, tokenizer) for service in services])
services_emb.shape
```

```python
test_query = embed_bert_cls('Какие документы необходимо предоставить для постановки на воинский учет?', model, tokenizer)

# в коде ниже мы получаем список из кортежей вида (индекс, dot<запрос, документ>)
# этот список мы сортируем по dot и возвращаем максимальный скор с индексом
test_index, test_score = max([(index, 1 - cosine(test_query, service)) for index, service in enumerate(services_emb)], key=lambda x: x[1])
test_index, test_score
```

```python
def make_query_dot(query: str):
    query_emb = embed_bert_cls(query, model, tokenizer)
    index, score = max(
        [
            (index, 1 - cosine(query_emb, service)) for index, service in enumerate(services_emb)
        ],
        key=lambda x: x[1]
    )
    print(f"QUERY: {query}")
    print(f"{score=}, {services[index]}")
    result = f"{query} {services[index]}"
    return result
```

```python
result_dot = make_query_dot('Какие документы необходимо предоставить для постановки на воинский учет?')
```

```python
for vk_query in vk_queries:
    make_query_dot(vk_query)
    print("\n========================================\n")
```

### **поиск `FAISS`**

- эмбеддер: `ruBERT-tiny2`
- [для небольших задач подойдёт `FlatL2`](https://github.com/facebookresearch/faiss/wiki/Guidelines-to-choose-an-index)

```python
%%capture
!pip install faiss-cpu
```

```python
import faiss
```

```python
index = faiss.IndexFlatL2(services_emb.shape[1])
```

```python
index.add(services_emb)
```

```python
def search_nearest_neighbors(index, query, k):
    scores, indices = index.search(np.array([query]), k)
    return scores, indices
```

```python
def make_query_faiss(query: str, k: int):
    query_emb = embed_bert_cls(query, model, tokenizer)
    scores, neighbors = search_nearest_neighbors(index, query_emb, k)
    print(f"QUERY: {query}")
    result = [query]
    for score, neighbor in zip(*scores, *neighbors):
        print(f"{score = }\n{services[neighbor]}")
        result.append(neighbor)
        print()
    return result
```
